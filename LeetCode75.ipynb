```python
def runningSum(nums):
  total=0
  arr = []
  for i in nums:
    total = total + i
    arr.append(total)
  return arr
```

```python
def runningSum(nums):
  total=0
  i=0
  arr = []
  while(i<len(nums)):
    total = total + nums[i]
    arr.append(total)
    i=i+1
  return arr
```

#724. Find Pivot Index


```python
def pivotIndex(nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        for i in range(len(nums)):
          if(sum(nums[0:i]) == sum(nums[i+1:len(nums)])):
              return i
        return -1
```

```python
def pivotIndex(newNums, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if(len(nums)!=0):
          if(sum(newNums[0:len(newNums)]) == sum(nums[1:len(nums)])):
            return len(newNums)
          else:
            newNums.append(nums[0])
            return pivotIndex(newNums, nums[1:len(nums)])
        else:
          return -1
```
